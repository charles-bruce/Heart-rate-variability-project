#Exploring thr relationship between physical activity (represennted by acceleration magnitude) and the HRV metric 

To explore the relationship between physical activity levels and HRV metrics, we will use the acceleration data as a proxy for physical activity level. A simple approach is to calculate the magnitude of acceleration (excluding gravitational acceleration ~9.81 m/s^2 to get activity-related accelerations) and then correlate this with the HRV metrics.



```python
df_female['Accel_Magnitude'] = np.sqrt(df_female['Shimmer_8663_Accel_LN_X_LPF_CAL']**2 +
                                                df_female['Shimmer_8663_Accel_LN_Y_LPF_CAL']**2 +
                                                df_female['Shimmer_8663_Accel_LN_Z_LPF_CAL']**2)

```

```python
df_male['Accel_Magnitude'] = np.sqrt(df_male['Shimmer_8663_Accel_LN_X_LPF_CAL']**2 +
                                                df_male['Shimmer_8663_Accel_LN_Y_LPF_CAL']**2 +
                                                df_male['Shimmer_8663_Accel_LN_Z_LPF_CAL']**2)

```

```python
# Calculate correlation between 'Accel_Magnitude' and 'Shimmer_8663_ECGtoHR_LL_RA_LPF_CAL' for both datasets
correlation_female = df_female['Accel_Magnitude'].corr(df_female['Shimmer_8663_ECGtoHR_LL_RA_LPF_CAL'])
correlation_male = df_male['Accel_Magnitude'].corr(df_male['Shimmer_8663_ECGtoHR_LL_RA_LPF_CAL'])

correlation_female, correlation_male
```

UNDERSTANDING HRV ACROSS ACTIVITIES


```python
# Box Plot for HRV Variability Across Activities
def plot_hrv_variability_boxplot(df, segments, title):
    # Prepare data for plotting: segment names and corresponding HRV values
    activities = []
    hrv_values = []
    for activity, (start, end) in segments.items():
        segment_data = df.iloc[start:end]
        activities.append(activity)
        hrv_values.append(segment_data['Shimmer_8663_ECGtoHR_LL_RA_LPF_CAL'].values)

    # Create box plot
    plt.figure(figsize=(10, 6))
    plt.boxplot(hrv_values, labels=activities)
    plt.title(title)
    plt.ylabel('HRV (BPM)')
    plt.xlabel('Activity')
    plt.xticks(rotation=45)
    plt.tight_layout()  # Adjust layout to make room for the label
    plt.show()
```

```python
# Plot HRV Variability for Female Participant
plot_hrv_variability_boxplot(df_female, activity_segments_female, 'HRV Variability Across Activities - Female Participant')

```

```python
# Plot HRV Variability for Male Participant
plot_hrv_variability_boxplot(df_male, activity_segments_male, 'HRV Variability Across Activities - Male Participant')
```

The box plots above display the variability in HRV (Heart Rate Variability) metrics across different activities (Resting, Walking, Jogging, Mentally Induced Stress) for both the female and male participants. These plots offer a visual representation of the spread and distribution of HRV values during each activity, highlighting:

Median HRV Rates: Indicated by the line within each box, showing the central tendency of HRV for each activity.

Variability and Range: The length of each box and the whiskers indicate the spread and range of HRV values, with longer boxes and whiskers signifying greater variability.

Outliers: Points outside the whiskers represent outlier HRV values, which could be of interest when monitoring for abnormal heart function.


Observations and Implications for Remote Monitoring:

Activity Impact on HRV: The plots demonstrate how HRV varies with activity type, with physical activities like walking and jogging generally showing higher variability compared to resting and mentally induced stress. This underscores the importance of activity context in interpreting HRV data for health monitoring.

Individual Differences: There are notable differences between the participants in terms of HRV response to the same activities, which emphasizes the need for personalized baselines in remote health monitoring systems.

Potential Alert Thresholds: The variability observed during each activity could inform the setting of dynamic alert thresholds in a monitoring system. For example, HRV values consistently below the lower quartile during resting could trigger an alert for further health assessment.



PATTERN RECOGNITION

This involves using machine learning to identify complex patterns in HRV data (and possibly other physiological parameters) that could indicate health issues beyond simple threshold breaches. This approach can uncover subtle changes in HRV patterns that precede health problems, potentially offering even earlier detection capabilities.

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Preparing the dataset for machine learning
# Adding activity labels based on segments
def label_activities(row, segments):
    for label, (start, end) in enumerate(segments.values()):
        if start <= row.name <= end:
            return label
    return -1  # Should not happen if segments cover all rows
```

```python
# Labeling the data
df_female['Activity'] = df_female.apply(label_activities, segments=activity_segments_female, axis=1)
df_male['Activity'] = df_male.apply(label_activities, segments=activity_segments_male, axis=1)

```

```python
# Combining datasets for a more robust model (optional based on your preference)
df_combined = pd.concat([df_female, df_male], ignore_index=True)
```

```python
# Selecting features and labels
X = df_combined[['Shimmer_8663_ECGtoHR_LL_RA_LPF_CAL']]  # Using HRV as the feature
y = df_combined['Activity']
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
 #Training a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
# Predicting on the test set
y_pred = clf.predict(X_test)
```

```python
# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=['Resting', 'Walking', 'Jogging', 'Mentally Induced Stress'])
```

```python
 #Define normal HRV thresholds based on resting data for a simplistic approach
resting_hrv_female = df_female[df_female['Activity'] == 0]['Shimmer_8663_ECGtoHR_LL_RA_LPF_CAL']
resting_hrv_male = df_male[df_male['Activity'] == 0]['Shimmer_8663_ECGtoHR_LL_RA_LPF_CAL']
resting_hrv_combined = pd.concat([resting_hrv_female, resting_hrv_male])

```

```python
lower_threshold, upper_threshold = resting_hrv_combined.quantile([0.25, 0.75])
```

```python
# Label the combined dataset as "Healthy" or "Potential Issue" based on HRV thresholds
df_combined['HRV_Status'] = df_combined['Shimmer_8663_ECGtoHR_LL_RA_LPF_CAL'].apply(
    lambda x: 'Healthy' if lower_threshold <= x <= upper_threshold else 'Potential Issue'
)
```

```python
# Prepare features and labels for the binary classification
X_binary = df_combined[['Shimmer_8663_ECGtoHR_LL_RA_LPF_CAL']]  # Features - HRV values
y_binary = df_combined['HRV_Status']  # Labels - 'Healthy' or 'Potential Issue'

```

```python
#Splitting the dataset into training and testing sets
X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(
    X_binary, y_binary, test_size=0.3, random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegressionCV
```

```python
# Logistic Regression with Cross-Validation and Regularization
clf_logistic_cv = LogisticRegressionCV(cv=5, penalty='l2', random_state=42, max_iter=1000, class_weight='balanced').fit(X_train_binary, y_train_binary)
```

```python
# Cross-validated scores on the training set
cv_scores = cross_val_score(clf_logistic_cv, X_train_binary, y_train_binary, cv=5)
```

```python
# Predicting on the test set
y_pred_logistic_cv = clf_logistic_cv.predict(X_test_binary)
```

```python
# Evaluating the model
accuracy_logistic_cv = accuracy_score(y_test_binary, y_pred_logistic_cv)
report_logistic_cv = classification_report(y_test_binary, y_pred_logistic_cv)
```

```python

print("Cross-validated scores:", cv_scores)
print("Test Accuracy:", accuracy_logistic_cv)
print(report_logistic_cv)
```
